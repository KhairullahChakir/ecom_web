# Phase 22: Exit Model (Model 2) - Deployment & Export
====================================================
Benchmarking the real-world Exit Model and exporting it to ONNX for the tracking engine.

```python
import torch
import numpy as np
import os
import sys

sys.path.append(os.path.abspath("../scripts"))
from train_tcn import AbandonmentTCN

MODEL_PATH = "../backend/models/exit_model_tcn.pth"
ONNX_OUTPUT = "../backend/models/exit_model_v1.onnx"

model = AbandonmentTCN(num_page_types=4)
model.load_state_dict(torch.load(MODEL_PATH, map_location='cpu'))
model.eval()

print("âœ… Loaded TCN model for export.")
```

```python
dummy_page = torch.randint(0, 4, (1, 20)).long()
dummy_dur = torch.randn(1, 20).float()

torch.onnx.export(
    model, 
    (dummy_page, dummy_dur),
    ONNX_OUTPUT,
    input_names=['page_ids', 'durations'],
    output_names=['logits'],
    dynamic_axes={'page_ids': {0: 'batch'}, 'durations': {0: 'batch'}},
    opset_version=12
)

print(f"ðŸš€ Successfully exported to: {ONNX_OUTPUT}")
```
